Eco Fusion Solutions: The three year EEE sustainability overview between 2023-2025

The northern lights project has allowed eco fusion to grow rapidly as a technology 
partner, yet with that growth comes added responsibility for environmental 
management. This technical report will allow a deep dive and strategic overview of the 
company's electrical and electronic equipment (EEE) data from 2023 to 2025 through 
summarizing yearly acquisition and disposal patterns and sustainability implications, 
while explaining the detailed category analysis will follow later in the report. The 
purpose of this report is to provide a concise, and visual overview of Eco Fusions
current sustainability performance and the trends that should inform the company's
next EEE strategies.

Across The different categories, the company has acquired roughly 20 000 EEE units 
through ought the three year peridot, while total disposals also increased each year. 
Through the years which have been examined there has been a steady inflow with small 
year to year changes, the disposal volumes rising in this case indicate the equipment 
turnover and maturing infrastructure. The recycling throughout the years forms a 
smaller proportion than expected when compared to the dataset meaning non recycled 
units exceed recycled ones every year. In order to visually show this figure 1.1 being a 
stacked bar chart compares the total acquired / recycled / non recycled per year 
revealing a clear sustainability gap.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))

Years = [2023, 2024, 2025]
Units_Acquired = [10887, 9174, 11498]
Units_Disposed_for_recycling = [536, 300, 371]
Units_Disposed_for_non_recycling = [3443, 2741, 3291]

x = np.arange(len(Years))
width = 0.15

plt.bar(x - width, Units_Acquired, width, label='Units Acquired')
plt.bar(x, Units_Disposed_for_recycling, width, label='Units Disposed for recycling')
plt.bar(x + width, Units_Disposed_for_non_recycling, width, label='Units Disposed for non-recycling')

plt.xlabel("Years")
plt.ylabel("Units")
plt.title("Sums per year per category")
plt.legend()

plt.xticks(x, Years)

plt.grid(True)
plt.tight_layout()
plt.show()
```

Year to year trends
From 2023 to 2025 the data shows uneven progress in sustainable disposal. Between 
2023 to 2024 acquisitions rose sharply, notably lamps and small equipment, but 
recycling rates fell despite this. Again between these years acquisition declined in most 
categories yet non – recycled disposals increased. Showing that total units acquired 
decreased by 15% yet non recyclable disposals grew 15 %. These data sets show that 
improved recycling efficiency has not yet matched the company’s scale of equipment 
renewal. Figure 1.2 visually display the comparison between total recycled vs non-recycled each year.

Between all the years examined recycling rates remain below 30%, which is considered 
to be well under what is normal seen as a sustainable performance for a ICT company.
In 2023 being 9.1% of all disposals being recycled, in 2024 7.6% despite lower total 
disposals and it 2025 10% of disposables being recycled being a small increase, yet still 
modest. This data confirms that 90% of waste in not recycled instead of reentering the 
material cycle. Furthermore data has revealed that temperature exchange equipment 
and small IT and telecom equipment are recycled the least. While lamps performed
better at 15 – 20 % being recycled. Overall, the company’s recycling system seems to 
be static, with performance fluctuating between 7 and 10 percent. This indicates 
structural rather than category-specific weaknesses in collection, sorting, and 
contractor follow-up.

Viewed from a strategic perspective, the three-year data highlights both operational 
growth and a widening sustainability gap. The Total EEE acquired = 26 820 units with the 
total being disposed reaching 9700 units. This equates to roughly 36% of all new 
acquisitions, suggesting shorter product lifespans or even limited reuse programs.
Further lack of “end of life management” is signified through the recycling lag of only 
870 units being recycled over three years against 8830 non-recycled. Such low recovery 
and recycling rates threaten compliance with EU WEEE directives, as these expect 
substantially higher recycling percentages for ICT equipment. Environmental this 
means unnecessary raw-material loss and carbon emissions, ultimately undermining
the company’s claim of being a sustainable Northern Lights partner.
